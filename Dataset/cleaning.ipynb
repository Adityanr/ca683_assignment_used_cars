```python
import pandas as pd

# Load the dataset
df = pd.read_csv('autos.csv')
```

```python
df
```

```python
# Remove any columns that have more than 50% of missing values
df = df.loc[:, df.isnull().mean() < 0.5]

# Drop any rows that have missing values
df = df.dropna()
df
```

```python
# Define a mapping of German values to English values
german_to_english = {
    'privat': 'private',
    'gewerblich': 'commercial',
    'Angebot': 'offer',
    'Gesuch': 'request',
    'limousine': 'sedan',
    'kleinwagen': 'compact car',
    'kombi': 'station wagon',
    'cabrio': 'convertible',
    'andere': 'other',
    'manuell': 'manual',
    'automatik': 'automatic',
    'benzin': 'gasoline',
    'diesel': 'diesel',
    'cng': 'natural gas',
    'lpg': 'liquefied petroleum gas',
    'notrepaired': 'not repaired',
    'ja': 'yes',
    'nein': 'no'
}

# Convert the specified columns from German to English
df['seller'] = df['seller'].map(german_to_english)
df['offerType'] = df['offerType'].map(german_to_english)
df['vehicleType'] = df['vehicleType'].map(german_to_english)
df['gearbox'] = df['gearbox'].map(german_to_english)
df['model'] = df['model'].map(german_to_english)
df['fuelType'] = df['fuelType'].map(german_to_english)
df['notRepairedDamage'] = df['notRepairedDamage'].map(german_to_english)
```

```python
df
```

```python
# Convert the Mileage, Power, Engine_Size, and Price columns to numeric types
df['kilometer'] = pd.to_numeric(df['kilometer'].replace('[^0-9]', ''))
df['powerPS'] = pd.to_numeric(df['powerPS'].replace('[^0-9]', ''))
# df['Engine_Size'] = pd.to_numeric(df['Engine_Size'].str.replace('[^0-9]', ''))
df['price'] = pd.to_numeric(df['price'].replace('[^0-9]', ''))
```

```python
df
```

```python
# Remove any rows where the Registration_Year is earlier than 1980 or later than 2021
df = df[(df['yearOfRegistration'] >= 1980) & (df['yearOfRegistration'] <= 2021)]
```

```python
df = df.dropna()
```

```python
df.to_csv('cleaned_used_cars.csv', index=False)
```
